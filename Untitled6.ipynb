Extracting image from video

```python
import cv2
import os

video_path = "myface.mp4.mp4"
output_folder = "/content/dataset/Om"

os.makedirs(output_folder, exist_ok=True)

cap = cv2.VideoCapture(video_path)

count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    cv2.imwrite(f"{output_folder}/img{count}.jpg", frame)
    count += 1

cap.release()

print("Images extracted:", count)
```

Loading the dataset

```python
!pip install face_recognition
import os
import face_recognition
import cv2

dataset_path = "/content/dataset"

known_encodings = []
known_names = []

for person in os.listdir(dataset_path):

    person_folder = os.path.join(dataset_path, person)

    for image_name in os.listdir(person_folder):

        image_path = os.path.join(person_folder, image_name)

        image = face_recognition.load_image_file(image_path)

        # resize image for faster processing
        small_image = cv2.resize(image, (0,0), fx=0.25, fy=0.25)

        encodings = face_recognition.face_encodings(small_image)

        if len(encodings) > 0:
            known_encodings.append(encodings[0])
            known_names.append(person)

print("Dataset Loaded Successfully")
```

```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode
import cv2
import numpy as np
from PIL import Image
import io

def take_photo(filename='photo.jpg', quality=0.8):
  js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);

      stream.getVideoTracks()[0].stop();
      div.remove();

      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')

  display(js)
  data = eval_js('takePhoto({})'.format(quality))
  binary = b64decode(data.split(',')[1])

  with open(filename, 'wb') as f:
    f.write(binary)

  return filename
```

```python
from datetime import datetime
import pandas as pd
import os
import face_recognition # Added import

subject = "Machine Learning"

now = datetime.now()
date = now.strftime("%Y-%m-%d")
time = now.strftime("%H:%M:%S")

# Define the expected columns for the attendance DataFrame
expected_columns = ["Name","Date","Time","Subject"]

# Check if the attendance file exists
if os.path.exists(attendance_file):
    # Read the existing CSV file
    df = pd.read_csv(attendance_file)

    # Check if the columns of the existing DataFrame match the expected columns
    if not list(df.columns) == expected_columns:
        print("Warning: The existing attendance.csv file has an incompatible column structure. It will be re-initialized with the correct columns. Old data will be lost.")
        df = pd.DataFrame(columns=expected_columns)
else:
    # If the file does not exist, create a new DataFrame with the expected columns
    df = pd.DataFrame(columns=expected_columns)

# --- Start of the fix: Incorporate face detection and name assignment ---
# Assuming 'image_path' (e.g., 'photo.jpg') is available from the previous take_photo() call
# and 'known_encodings', 'known_names' are loaded from the dataset.

image_to_check = face_recognition.load_image_file(image_path)
face_locations_check = face_recognition.face_locations(image_to_check)
face_encodings_check = face_recognition.face_encodings(image_to_check, face_locations_check)

name_to_record = "Unknown" # Initialize name to a default value

if face_encodings_check: # Check if any faces were detected
    for face_encoding_single in face_encodings_check:
        matches = face_recognition.compare_faces(known_encodings, face_encoding_single)

        # Find the matched name. For simplicity, we'll take the first match.
        if True in matches:
            index = matches.index(True)
            name_to_record = known_names[index]
            break # Assuming we want to record attendance for one person per photo
else:
    print("No faces detected in the captured photo.")
# --- End of the fix ---

# Append the new attendance record
df.loc[len(df)] = [name_to_record, date, time, subject]

# Save the updated DataFrame back to the CSV file
df.to_csv(attendance_file,index=False)

print(f"Attendance for {name_to_record} on {date} at {time} for {subject} marked.")
```
